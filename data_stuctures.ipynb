# **Data Structures - List, Dictionary, Set & Conditional Statements** *

# 1. List Creation:





```python

# a.  Create a list named age_list with five integer elements.
# b. Create a list named name_list with five string elements.

age_list=[20,25,27,30,23]
name_list=["saranya","sanjith","priya","arun","mani"]
print(f"age list :{age_list}")
print(f"name list is :{name_list}")

# 2. List Operations / Modifications:
# a. Append the string "Yazhini" to name_list.

name_list.append("yazhini")
print(f"name list is after append :{name_list}")

# b.  Insert the element 26 at index 2 in age_list
age_list.insert(2, 26)
print(f"age list after insert:{age_list}")

#Remove the string "Yazhini" from name_list.
name_list.remove("yazhini")
print(f"name list is after remove :{name_list}")

#Pop the last element from age_list.
age_list.pop()
print(f"age list after pop:{age_list}")

#  Extend the age_list with additional ages [31,30,32]
age_list.extend([31,30,32])
print(f"age list after extend:{age_list}")

#Sort age_list in descending order.
age_list.sort(reverse=True)
print(f"age list after sort:{age_list}")

#Find the max age, min age, and sum of all ages from age_list.
max_age = max(age_list)
print(f"maximum age is:{max_age}")
min_age = min(age_list)
print(f"minimum age is:{min_age}")
sum_age = sum(age_list)
print(f"sum ofage is:{sum_age}")

# 3. Accessing List Elements:
#a. Print the first element of name_list.

print(f"first element of name list is:{name_list[0]}")

#b.  Print the last element of name_list.
print(f"last element of name list is:{name_list[-1]}")

#c.Print the elements from index 2 to index 4 in name_list.
print(f"element of index 2 to 4 is:{name_list[2:5]}")

#d.  Print the elements of name_list in reverse order.
print(f"reverse order of name list is:{name_list[::-1]}")
```

## dictionary (Creation, Modification and Access):  

```python

# a. Create a dictionary named student_marks that maps  to their marks
student_marks = {"saranya": 76,"arun": 85,"priya": 92,"divya": 67,"mani": 74}

print(f"student and marks is:{student_marks}")

#Access and print the mark of a specific .
print(f"mark of priya: {student_marks["priya"]}")

#  Add a new student, "Janani," with a mark of 80 to the student_marks dictionary.
student_marks["janani"] = 80
print(f"student mark after add: {student_marks}")

#. Update the mark of any one older student to 82.
student_marks["saranya"] = 82
print(f"student mark after update: {student_marks}")

# print all keys, values, and key-value pairs in the student_marks dictionary.
print(f"all keys: {student_marks.keys()}")
print(f"all values:{ student_marks.values()}")
print(f"all items: {student_marks.items()}")
```

# Sets (Operations):  

```python
# creating set
my_set={"a","e","i","o","u","a","a","i"}
print (f"my set is:{my_set}")
# A set in programming is a data structure that stores unique elements only. That means it does not allow duplicates.

# b. Attempt to access my_set[4]. If the code throws an error, provide an explanation.
# print(f"fourth element of set:{my_set[4]}")
# An error occurs because sets in Python do not support indexing.

# c. create two sets
set1 = {1, 3, 5, 7, 9}
set2 = {2, 3, 5, 8, 10}

# d. compute union and intersection
union_result = set1.union(set2)
intersection_result = set1.intersection(set2)

print(f"union of set1 and set2:{ union_result}")
print(f"intersection of set1 and set2:{intersection_result}")
```

# Operators & Conditional Statements : (IF, ELIF, ELSE)  

```python
# 1.  user for input
score = int(input("enter your score : "))

# 2. find the performance category
if score > 7:
    print("above average: excellent work, keep it up.")
elif 4 <= score <= 7:
    print("average: good effort, keep practicing, there's room for improvement.")
else:
    print("below average: need to improve your performance, consistent practice will lead to better results.")
```
